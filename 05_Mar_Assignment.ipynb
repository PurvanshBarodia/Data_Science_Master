Q1. How can you create a Bokeh plot using Python code?

```python
import bokeh.io
import bokeh.plotting
bokeh.io.output_notebook()
```

```python
# Step 1: import necessary modules
from bokeh.plotting import figure, output_file, show

# Step 2: Prepare the data
x = [1, 2, 3, 4, 5]
y = [6, 7, 2, 4, 5]

# Step 3: Create the plot
p = figure(title='My Bokeh Plot', x_axis_label='X', y_axis_label='Y')
p.circle(x, y, size=10, color='red', alpha=0.5)

# Step 4: Specify the output file
output_file('my_plot.html')

# Step 5: Display the plot
show(p)
```

Q2. What are glyphs in Bokeh, and how can you add them to a Bokeh plot? Explain with an example.

In Bokeh, glyphs are used to create various types of visualizations, such as scatter plots, line plots, bar charts. Here is the example

```python
x = [1, 2, 3, 4, 5]
y = [6, 7, 2, 4, 5]

p = figure(title='My Bokeh Plot', x_axis_label='X', y_axis_label='Y')

p.square(x, y, size=12, color='green', alpha=0.8)
p.line(x, y, line_width=2, color='blue')

output_file('my_plot.html')

show(p)
```

Q3. How can you customize the appearance of a Bokeh plot, including the axes, title, and legend?

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
from bokeh.plotting import figure, output_file, show

x = [1, 2, 3, 4, 5]
y = [6, 7, 2, 4, 5]

p = figure(title='My Bokeh Plot', x_axis_label='X', y_axis_label='Y')

p.circle(x, y, size=10, color='red', alpha=0.5)
p.line(x, y, line_width=2, color='blue')

p.xaxis.axis_label = "X-axis label"
p.yaxis.axis_label = "Y-axis label"
p.xaxis.major_label_text_color = "green"
p.yaxis.major_label_text_font_size = "12pt"

p.title.text = "Customized Bokeh Plot"
p.title.text_font = "Arial"
p.title.text_font_size = "18pt"
p.title.align = "center"

p.legend.title = "Legend"
p.legend.location = "top_right"
p.legend.label_text_font_size = "12pt"
p.legend.label_text_color = "navy"

output_file('my_plot.html')

show(p)
```
